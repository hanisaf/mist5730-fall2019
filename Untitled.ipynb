```python
import altair as alt
import pandas as pd
```

```python
data = pd.read_csv('mpg.csv')
```

```python
data.head()
```

```python
data2 = data.groupby('cylinders')['mpg'].mean().reset_index()
data2.columns = ['cylinders', 'Average of mpg']
```

```python
chart2 = alt.Chart(data).mark_bar().encode(
    x='average(mpg)',
    y='cylinders:N'
)
```

```python
chart1 = alt.Chart(data2).mark_bar().encode(
    x='Average of mpg',
    y='cylinders:N'
)
```

```python
alt.renderers.enable('notebook')
```
