This notebook uses the [GapMinder data set](https://www.gapminder.org)

```python
import pandas as pd
data = pd.read_csv("data/gapminder.tsv", sep='\t')
data.head() # show five entries
```

Answer the following questions using indexing operations
- Create a table which rows are `country` and columns are `year`. Values are the country's `pop` in that year
- What is the average lifeExp per year
- What is the average gdpPercap per continent


```python
s=data.set_index(['country','year'])['pop'].unstack('year')
s
```

```python
s.mean()
```

```python
data.set_index(['country','year'])['lifeExp'].unstack('year').mean()
```

```python
data.set_index(['country', 'continent', 'year'])['gdpPercap'].unstack('continent').mean()
```
