This notebook uses the [GapMinder data set](https://www.gapminder.org)

```python
import pandas as pd
data = pd.read_csv("data/gapminder.tsv", sep='\t')
data.head() # show five entries
```

Using multi-indexing, answer the following questions
- Select data about countries in Asia 
- Select data about countries in Asia in years 1952 to 2000
- What is the average `pop` of countries in Asia in years 1952 to 2000

```python
data.set_index('continent').loc['Asia']
```

```python
asian_countries = data.set_index('continent').loc['Asia']
mask = (asian_countries['year'] >= 1952) & (asian_countries['year'] <= 2000)
asian_countries_1952_2000 = asian_countries[mask]
asian_countries_1952_2000
```

```python
asian_countries_1952_2000['pop'].mean()
```
