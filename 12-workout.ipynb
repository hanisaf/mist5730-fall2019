The goal of this notebook is to introduce Pandas using the [Gapminder data](https://www.gapminder.org/data/)

A copy of the data is available in this [code repo](./data/gapminder.tsv)

1- Load the dataset in the notebook

```python
import pandas as pd
data = pd.read_csv("data/gapminder.tsv", sep='\t')
data.head() # show five entries
```

2- How many rows and cols does the data contain?

```python
(len(data.index), len(data.columns))
```

3- Select the country column from the dataset

```python
# either data['country'] or data.country
data['country']
```

4- select country and continent

```python
data[['country', 'continent']]
```

5- select row 1

```python
# if row 1 means the first row
row1 = data.iloc[0]
row1
```

```python
# if row 1 means the row of index 1
row1 = data.loc[1]
row1
```

6- select rows 1 to 4

```python
# again, if we are extracting the first four rows
rows = data.iloc[0:5]
rows
```

```python
# but if we are extracting rows of indicies 1,2,3,4
# notice that the slicing is inclusive
rows = data.loc[1:4]
rows
```
