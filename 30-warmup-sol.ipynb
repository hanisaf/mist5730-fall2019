```python
import pandas as pd
import altair as alt
alt.renderers.enable('notebook')
```

1- Load the Gapminder dataset

```python
data = pd.read_csv('data/gapminder.tsv', sep='\t')
```

```python
data.head()
```

2- Relate per capita GDP to life expectancy using point encoding

```python
alt.Chart(data).mark_point().encode(    
    x = 'gdpPercap',
    y = 'lifeExp')
```

2- Use color to differentiate between the different years

```python
alt.Chart(data).mark_point().encode(    
    x = 'gdpPercap',
    y = 'lifeExp',
    color = 'year:O',
    tooltip='country'
)
```

4- Can you account for population size? Try changing the size of the dots

```python
alt.Chart(data).mark_point().encode(    
    x = 'gdpPercap',
    y = 'lifeExp',
    color = 'year:O',
    tooltip='country',
    size=alt.Size('pop', scale = alt.Scale(range=(50,1500)))
)
```

```python
alt.Scale(range=(50,150))
```

5- Use facets (cols/rows) to break this plot into one plot per year

```python
alt.Chart(data).mark_point().encode(    
    x = 'gdpPercap',
    y = 'lifeExp',
    row = 'year',
    tooltip='country',
    size=alt.Size('pop', scale = alt.Scale(range=(50,1500)))
)
```

6- how does life expectancy vary per continent?

```python
alt.Chart(data).mark_bar().encode(    
    y = 'continent',
    x = 'min(lifeExp)',
    x2 = 'max(lifeExp)',
    tooltip = 'average(lifeExp)'
)
```

7- how did life expectancy change over time?

```python
alt.Chart(data).mark_bar().encode(    
    y = 'year:O',
    x = 'min(lifeExp)',
    x2 = 'max(lifeExp)',
    tooltip = 'average(lifeExp)',
) + alt.Chart(data).mark_point(color='red').encode(    
    y = 'year:O',
    x = 'average(lifeExp)',
    tooltip = 'average(lifeExp)',
)
```
